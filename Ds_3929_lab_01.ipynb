## Explaination:  **MERGE_DATASET.isnull().sum()** is used to check null vlaues andd **.fillna(_)** method is used to fill up null values . As we check that Cpu_frequency have float datatype so it's a numerical data and Manufacturer is of object type so it is categorical data set so we use mean,mode according to the prescribed instructions

```python
MERGE_DATASET.isnull().sum()
```

```python
missing_percent = MERGE_DATASET.isnull().sum() / len(MERGE_DATASET) * 100
missing_percent
```

```python
# DROP UNNMAED 5 AND 6 AS THEY'VE>50% MISSING VALUES

cols_to_drop = missing_percent[missing_percent > 50].index
MERGE_DATASET = MERGE_DATASET.drop(columns=cols_to_drop)



```

```python
MERGE_DATASET.head()
```

```python
MERGE_DATASET.isnull().sum()
```

```python
MERGE_DATASET.info()
```

```python
# As CPU_Frequecy column is numerical so we've to fill it with either median or mode

mean_val=MERGE_DATASET['CPU_frequency'].mean()
mean_val
```

```python
# now fill it with the missing values in the columns

MERGE_DATASET['CPU_frequency'].fillna(mean_val,inplace=True)

MERGE_DATASET.isnull().sum()
```

```python
# get random 5 rows each time

MERGE_DATASET.sample(n=5)
```

```python
mode_value=MERGE_DATASET['Manufacturer'].mode()[0]
mode_value
```

```python
MERGE_DATASET['Manufacturer'].fillna(mode_value,inplace=True)

```

```python
MERGE_DATASET.isnull().sum()
```

```python
MERGE_DATASET.info()
```

### ***TASK#03***

pandas give us cut() method to create new labeled columns and create bins and labels also and for summary we use describe() to get full detail about the dataset

```python
MERGE_DATASET['Price'].sample(n=5)
```

```python
# suppose  low=0-1000,  medium=1001-2000  ,  high= price>2000


create_bins=[0,1000,2000, float('inf')]
bins_labels=['low','medium','high']

MERGE_DATASET['price_category']=pd.cut(MERGE_DATASET['Price'],bins=create_bins,labels=bins_labels)

MERGE_DATASET.head()
```

```python

from sklearn.preprocessing import MinMaxScaler

numerical_features = ['Screen_Size_inch', 'CPU_frequency', 'RAM_GB', 'Weight_pounds']
scaler = MinMaxScaler()
MERGE_DATASET[numerical_features] = scaler.fit_transform(MERGE_DATASET[numerical_features])
```
